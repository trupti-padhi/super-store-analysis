## **Dataset description**

Here it is a super store data set,Having columns 'Row ID', 'Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State','Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category','Product Name', 'Sales', 'Order Month', 'Order Year', 'Days to Ship'.


```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## lets see if the dataset consist any duplicate data or not if there is any duplicate data lets clean it
sale_df.drop_duplicates(inplace=True)
uniq_row_numbers=sale_df.shape[0]
uniq_row_numbers ### output is 87359 that means the duplicate valu are deleated
```

```python
null_value=sale_df.isnull()==True# Missing Values/Null Values Count
sale_df.fillna(np.nan,inplace=True)
sale_df.isnull().sum()
```

```python
rows_with_null = sale_df[sale_df['Postal Code'].isnull()]
rows_with_null
# Fill null values in 'Postal Code' with '05407'
sale_df['Postal Code'].fillna(5407, inplace=True)
sale_df.isnull().sum()
```

```python
sale_df['Postal Code'] = sale_df['Postal Code'].astype(int)
sale_df.info()
```

```python
# Convert 'Order Date' and 'Ship Date' to datetime dtype
sale_df['Order Date'] = pd.to_datetime(sale_df['Order Date'], format='%d/%m/%Y') # Specify the correct format
sale_df['Ship Date'] = pd.to_datetime(sale_df['Ship Date'], format='%d/%m/%Y') # Apply the same format for 'Ship Date'
```

```python
# Extract month and year into new columns
sale_df['Order Month'] = sale_df['Order Date'].dt.month
sale_df['Order Year'] = sale_df['Order Date'].dt.year
```

```python
# Calculate the difference in days between 'Ship Date' and 'Order Date'
sale_df['Days to Ship'] = (sale_df['Ship Date'] - sale_df['Order Date']).dt.days
```

# **Data visualization**

```python
# Count orders per year
orders_per_year = sale_df['Order Year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
orders_per_year.plot(kind='bar', color='skyblue')
plt.title('Orders Count per Year')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Here we found that we have highest number of orders in year 2018

## **Highest revenue generation column wise**

```python
# Aggregate sales by year
revenue_per_year = sale_df.groupby('Order Year')['Sales'].sum()

# Find the year with the highest revenue
year_highest_revenue = revenue_per_year.idxmax()
highest_revenue = revenue_per_year.max()

# Plotting revenue per year
plt.figure(figsize=(10, 6))
revenue_per_year.plot(kind='bar', color='skyblue')
plt.title('Revenue per Year')
plt.xlabel('Year')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.annotate(f'Highest Revenue: {highest_revenue:.2f}', xy=(year_highest_revenue, highest_revenue), xytext=(year_highest_revenue, highest_revenue * 1.05),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             )
plt.show()
```

Here we found that we have generated highest revenue in year 2018

```python
# Aggregate sales by segment
segment_revenue = sale_df.groupby('Segment')['Sales'].sum()

# Plotting revenue per segment as a pie chart
plt.figure(figsize=(8, 6))
plt.pie(segment_revenue, labels=segment_revenue.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Revenue by Segment')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

From the above chat we found that consumer helps us in generating highes revenue

```python
# Aggregate sales by category
category_revenue = sale_df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

# Plotting revenue per category as a bar plot
plt.figure(figsize=(8, 6))
plt.bar(category_revenue.index, category_revenue.values, color='skyblue')
plt.title('Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

Here we found that in Technology category we have generated highest brevenue

```python
# Aggregate sales by city
city_revenue = sale_df.groupby('City')['Sales'].sum().sort_values(ascending=False)

# Select top 10 cities with highest revenue
top_10_cities = city_revenue.head(10)

# Plotting top 10 cities by revenue as a bar plot
plt.figure(figsize=(12, 6))
top_10_cities.plot(kind='bar', color='skyblue')
plt.title('Top 10 Cities by Revenue')
plt.xlabel('City')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

From the above graph we found that New york city has the highes number of orderd

```python
# Aggregate sales by state
state_revenue = sale_df.groupby('State')['Sales'].sum().sort_values(ascending=False)

# Select top 10 states with highest revenue
top_10_states = state_revenue.head(10)

# Plotting top 10 states by revenue as a bar plot
plt.figure(figsize=(12, 6))
top_10_states.plot(kind='bar', color='skyblue')
plt.title('Top 10 States by Revenue')
plt.xlabel('State')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

From the above graph we found that california state has the highest number of orders

```python
# Aggregate sales by region
region_revenue = sale_df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

# Plotting revenue distribution by region as a pie plot
plt.figure(figsize=(8, 8))
plt.pie(region_revenue, labels=region_revenue.index, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'lightskyblue', 'lightpink'], startangle=140)
plt.title('Revenue Distribution by Region')
plt.show()
```

From the above chart we found that west regeion helps us in generating highest amount of revenue

```python
# Count the frequency of each segment in sale_df
segment_counts = sale_df['Segment'].value_counts()

# Plotting segment preferences as a pie chart
plt.figure(figsize=(8, 6))
plt.pie(segment_counts, labels=segment_counts.index, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Segment Preferences')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

From the above graph we found that the most valuable segment for us is consumer

# **Shipment time analysis**

```python
# Average and median shipping times
average_shipping_time = sale_df['Days to Ship'].mean()
median_shipping_time = sale_df['Days to Ship'].median()
print(f"Average Shipping Time: {average_shipping_time}")
print(f"Median Shipping Time: {median_shipping_time}")
```

Here we found that the most prefered shipment time is 3 to 4 days

```python
# Shipping times by Ship Mode
shipping_by_mode = sale_df.groupby('Ship Mode')['Days to Ship'].describe()
print(shipping_by_mode)
```

From the above we found that first class ship mode take 2 day to ship
same day ship mode take half day to ship
second class ship mode take 3 day to ship and
standard clas ship mode take 5 day to ship

```python
# Visualization: Shipping times by Ship Mode
plt.figure(figsize=(12, 6))
sns.boxplot(x='Ship Mode', y='Days to Ship', data=sale_df)
plt.title('Shipping Times by Ship Mode')
plt.show()
```

#**Customer Insights**

```python

# Segment customers based on Segment
customer_segment_analysis = sale_df.groupby('Segment')['Sales'].describe()
print(customer_segment_analysis)
```

```python
# Impact of shipping times on customer satisfaction (approximated by repeat purchases)
repeat_customers = sale_df.groupby('Customer ID').filter(lambda x: len(x) > 1)
repeat_customer_analysis = repeat_customers.groupby('Customer ID')['Days to Ship'].mean()
print(repeat_customer_analysis.describe())
```

Most repeat customers receive their orders within 4 to 4.5 days.
There’s a small variation in shipping times, indicating consistency.
A few outliers exist with exceptionally fast or slow shipping times, but these are rare.
These insights suggest that the shipping process is generally reliable for repeat customers, with room for slight improvement in consistency.

```python
# Correlation between Sales and Days to Ship
sales_ship_corr = sale_df[['Sales', 'Days to Ship']].corr()
print(sales_ship_corr)
```

Faster shipping times might have a very slight positive impact on sales, but the effect is not strong.
Other factors likely play a more significant role in influencing sales.
These insights suggest that while reducing shipping times could marginally improve sales, focusing on other aspects (e.g., product quality, customer service) might yield better results.

```python
# Visualization: Sales vs. Days to Ship
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Days to Ship', y='Sales', data=sale_df)
plt.title('Sales vs. Days to Ship')
plt.show()
```

The plot reinforces that shipping time has minimal influence on sales performance.
Other Factors: Other factors besides shipping duration likely have a more significant impact on sales figures.

```python
# Seasonal trends
sale_df['Order Month'] = sale_df['Order Date'].dt.month
sale_df['Order Year'] = sale_df['Order Date'].dt.year
seasonal_sales = sale_df.groupby(['Order Year', 'Order Month'])['Sales'].sum().reset_index()
print(seasonal_sales)
```

```python
import matplotlib.pyplot as plt

# Assuming you already have seasonal_sales DataFrame with Order Year, Order Month, and Sales columns

# Plotting the data
plt.figure(figsize=(10, 6))  # Adjust the figure size if needed

# Extracting data for plotting
years = seasonal_sales['Order Year']
months = seasonal_sales['Order Month']
sales = seasonal_sales['Sales']

# Creating a list of x-axis labels in 'Month Year' format for better readability
month_year_labels = [f"{month}-{year}" for year, month in zip(years, months)]

plt.plot(month_year_labels, sales, marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Month Year')
plt.ylabel('Sales')
plt.title('Seasonal Sales Trends')

# Rotating x-axis labels for better readability if there are many data points
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()  # Ensures all elements fit nicely in the plot area
plt.grid(True)  # Add grid for better readability
plt.show()
```

```python
# Visualization: Seasonal Sales Trends
plt.figure(figsize=(12, 6))
sns.lineplot(x='Order Month', y='Sales', hue='Order Year', data=seasonal_sales)
plt.title('Seasonal Sales Trends')
plt.show()
```
